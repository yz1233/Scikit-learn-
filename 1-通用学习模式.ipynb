```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

import numpy与sklearn数据集
用到sklearn两个模块（之后会讲,cross_validation过老已更新为model_selection），一个是划分训练集与测试集，一个是K最近邻法分类

```python
iris = datasets.load_iris()
iris_X = iris.data
iris_Y = iris.target
print(iris_X[:2,:])
```

iris为datasets中有的一种花（load导入），data中存放花的所有属性，target存放花的分类
运行得到结果X有四种属性

```python
print(iris_Y)
```

Y为三种分类

```python
X_train, X_test, Y_train, Y_test = train_test_split(iris_X, iris_Y, test_size=0.3)
print(Y_train)
```

划分训练集与测试集，测试集大小为0.3
运行结果会打乱数据顺序，乱的数据更利于机器学习

```python
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
print(knn.predict(X_test))
print(Y_test)
```

knn选择运用的分类器模块
knn.fit输入训练参数，自动完成训练工作
print输出训练完的knn预测值与测试集真实值
