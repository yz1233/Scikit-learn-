```python
from sklearn.model_selection import validation_curve  #可视化学习的过程
from sklearn.datasets import load_digits  #数字data
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
```

导入上节的代码，只有learning_curve变为validation_curve

```python
digits=load_digits()
X=digits.data
Y=digits.target
param_range=np.logspace(-6,-2.3,5)  #从-6log和-2.3log间取5个线段端点作为变量的变化
train_loss,test_loss=validation_curve(
    SVC(),X,Y,param_name='gamma',param_range=param_range,cv=10,scoring='neg_mean_squared_error')
train_loss_mean=-np.mean(train_loss,axis=1)
test_loss_mean=-np.mean(test_loss,axis=1)
```

同上，多了一个去掉train_sizes的步骤,加上了parameter的变化，validation的参数跟着变化，多了变量名和变量变化，去掉了SVC内的变量名

```python
plt.plot(param_range,train_loss_mean,'o-',color='r',label='Training')
plt.plot(param_range,test_loss_mean,'o-',color='g',label='Cross-validation')
plt.xlabel('gamma')
plt.ylabel('Loss')
plt.legend(loc='best')
plt.show()
```

此处与上节基本相同相同，改变了参数
