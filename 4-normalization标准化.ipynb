```python
from sklearn import preprocessing
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

例行导入数据模块和预处理，sklearn.datasets.samples_generator已被废除，可以直接写成sklearn.datasets，导入分类的数据集和SVM分类器

```python
a = np.array([[10,2.7,3.6],
              [-100,5,-2],
              [120,20,40]],dtype=np.float64)
print(a)
print(preprocessing.scale(a))
```

创建一个三维数组a（假设每一列是一个feature），输出a和预处理后的a（normalization又叫scale）进行对比

```python
X,Y = make_classification(n_samples=300,n_features=2,n_redundant=0,n_informative=2,random_state=22,n_clusters_per_class=1,scale=100)
```

生成数据，有300个data，2个特征，0个冗余特征个数，2个有信息的特征，固定随机数种子22，每个类别包含的的簇（聚集点）数量即每一类有一个聚集点，数据缩放比例为100倍

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show
```

plot出来看一下，c=Y表示按照Y的值划分颜色，两个特征分别做横纵坐标

```python
X = preprocessing.minmax_scale(X,feature_range=(-1,1))
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.3)
clf = SVC()
clf.fit(X_train,Y_train)
print(clf.score(X_test,Y_test))
```

minmax可设置scale的范围，默认0，1，此时设置为-1，1
划分训练测试集，定义分类器，将参数带入，打印最终得分
